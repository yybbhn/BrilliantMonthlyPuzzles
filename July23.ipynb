```python
BRILLIANT JULY 2023 PUZZLE
==========================


_______________
STATED PROBLEM:
Congratulations! You won free one-way tickets for life on Somewhere Air. The only problem? Each trip, your destination is chosen at random.
Somewhere Air has service to 25 cities, including yours. Each city has direct flights to every other city. Once you begin jet-setting, what's the average number of flights you’ll take before you get back home?


_____________
My solution:
Any given flight will take me from 1 of the 25 cities to 1 of the other 24 cities. This means any given flight has a 1 in 24 chance of taking me home. This occur on the first flight, or not at all if I'm so unlucky. 

Since each location has direct flights to the other 24 cities, we can call each ticket's destination independent of the others. Each flight can be defined as a success or failure, dependent on whether or not we get home. Finally, the probability of success is 1 in 24 for any given flight. With these 3 statements, we can use the geometric distribution to model this problem.

For a geometric distribution, the average number of trials (flights) until a success is the inverse of our probability. E(X) = 1/p = 1/(1/24)... 24 flights on average.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We'll now simulate this problem, incrasing the number of iterations in each go
# We should expect our average to get closer to 24 as we increase the number of iterations

T = 9
trys = np.zeros(T)
for t in range(T):
    n = 0
    i = 0
    while i != 1: #any integer from 1 to 24 inclusive can be used here as the "correct" destination
        n += 1
        i = np.random.choice(np.arange(1,25)) #25 is not included so this is only 1 through 24 inclusive
        trys[t] = n

avg0 = np.average(trys)
plt.plot(np.arange(0, T),trys)
plt.axhline(y = np.average(trys), color = 'r',  linestyle = ':', label = f'Avg = {np.average(trys)}')
plt.xlabel('Simulation No.')
plt.ylabel('Flights Taken')
plt.legend();
```
